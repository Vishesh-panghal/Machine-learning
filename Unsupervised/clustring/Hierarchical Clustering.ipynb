## Hierarchical Clustering

## <span style='color:yellow'> **Algorithm**. </span>
1. Initilize the Promixity matrix.
2. Make each point a cluster.
3. Inside a loop
   - Merge the 2 closest cluster.
   - Update the Promixity matrix.
4. Until only one cluster is left.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../Datasets/shoppingdata.csv")
df.head(5)
```

```python
data = df.iloc[:,3:5].values
data
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,8))
plt.title("Dendogram")
dend = sch.dendrogram(sch.linkage(data,method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=5,metric = 'euclidean',linkage='ward')
lables_ = cluster.fit_predict(data)
lables_
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data[:,0],data[:,1],c= cluster.labels_,cmap='rainbow')
plt.show()
```
