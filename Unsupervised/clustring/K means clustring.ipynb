## K means clustring

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
from sklearn import datasets
from sklearn.metrics import confusion_matrix,silhouette_score
```

```python
df = datasets.load_iris()
```

```python
X = df.data
y = df.target
```

```python
names = df.feature_names
names
```

```python
X,y = shuffle(X,y,random_state=42)
```

```python
km = KMeans(n_clusters=3,random_state=42)
iris_km = km.fit(X)
iris_km
```

```python
iris_km.labels_
```

```python
y = np.choose(y,[1,2,0]).astype(int)
y
```

```python
cf = confusion_matrix(y,iris_km.labels_)
cf
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1, 2, 1,)
ax.matshow(cf, cmap=plt.cm.Blues, alpha=0.3)
for i in range(cf.shape[0]):
    for j in range(cf.shape[1]):
        ax.text(x=j, y=i,s=cf[i, j], va='center', 
                ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(X[:, 3], X[:, 0], X[:, 2], 
            c=iris_km.labels_.astype(float), 
           edgecolor="k", s=150, cmap='viridis')
ax1.view_init(20, -50)
ax1.set_xlabel(names[3], fontsize=12)
ax1.set_ylabel(names[0], fontsize=12)
ax1.set_zlabel(names[2], fontsize=12)
ax1.set_title("K-Means Clusters for the Iris Dataset", fontsize=12)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')

for label, name in enumerate(['virginica','setosa','versicolor']):
    ax2.text3D(
        X[y == label, 3].mean(),
        X[y == label, 0].mean(),
        X[y == label, 2].mean() + 2,
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.2, edgecolor="w", facecolor="w"),
    )

ax2.scatter(X[:, 3], X[:, 0], X[:, 2], 
            c=y, edgecolor="k", s=150, 
            cmap='viridis')
ax2.view_init(20, -50)
ax2.set_xlabel(names[3], fontsize=12)
ax2.set_ylabel(names[0], fontsize=12)
ax2.set_zlabel(names[2], fontsize=12)
ax2.set_title("Actual Labels for the Iris Dataset", fontsize=12)
plt.show()
```

## Elbow method:- 

```python
inertia = []
for n_clusters in range(1, 11):
    km = KMeans(n_clusters=n_clusters, random_state=42)
    km.fit(X)
    inertia.append(km.inertia_)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

## Silhouette method

```python
silhouette_scores = []
cluster_range = range(2, 11)  # Silhouette score cannot be calculated for cluster = 1
for k in cluster_range:
    km = KMeans(n_clusters=k, random_state=42)
    cluster_labels = km.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Various Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```
